# Analys av Naturvårdsverkets bidrag för lokala naturvårdssatsningar (LONA)

av [Anna Norberg](www.twitter.com/annanorberg)  

Grunddata har begärts ut från Naturvårdsverket. Resultatet från analysen har använts som underlag till en artikel som skickades ut till Nyhetsbyrån Sirens kunder den 4 september 2019.  

```r
library(readxl)
library(here)
library(dplyr)
library(xlsx)
library(tidyr)
library(openxlsx)
```

**Läs in filer.**

```r
df <- read_excel(here("raw_data", "lona.xlsx"))
head(df)
```

**Undersök kolumnnamn och välj ut kolumner.**

```r
colnames(df)
```

```r
df_lona <- df %>%
  select(Län, Kommun, Ansökningsår, `Summa Beviljat Bidrag`) %>% 
  filter(Ansökningsår < 2019)
head(df_lona, 20)
```

**Hur mycket LONA-stöd har betalats ut totalt varje år?**

**Summera per år för hela riket.**

```r
year <- df_lona %>% 
  group_by(Ansökningsår) %>% 
  summarize("Totalt beviljat (kr)"=sum(`Summa Beviljat Bidrag`, na.rm=TRUE))
head(year, 10)
```

**Hur ser det ut på länsnivå? Summera per län och år.**

```r
lan_peryear <- df_lona %>% 
  group_by(Län, Ansökningsår) %>% 
  summarize("Totalt beviljat (kr)"=sum(`Summa Beviljat Bidrag`, na.rm=TRUE)) %>% 
  spread(Ansökningsår, `Totalt beviljat (kr)`)
```

```r
head(lan_peryear)
```

```r
lan_tot <- df_lona %>% 
  group_by(Län) %>% 
  summarize("Totalt beviljat (kr)"=sum(`Summa Beviljat Bidrag`, na.rm=TRUE))
lan <- left_join(lan_peryear, lan_tot, by="Län")
```

```r
head(lan)
```

**Hur mycket LONA-bidrag har kommunerna fått varje år?**

**Summera per kommun och år.**

```r
kommun_peryear <- df_lona %>% 
  group_by(Län, Kommun, Ansökningsår) %>% 
  summarize("Totalt beviljat (kr)"=sum(`Summa Beviljat Bidrag`, na.rm=TRUE)) %>% 
  spread(Ansökningsår, `Totalt beviljat (kr)`)
```

```r
kommun_tot <- df_lona %>% 
  group_by(Kommun) %>% 
  summarize("Totalt beviljat (kr)"=sum(`Summa Beviljat Bidrag`, na.rm=TRUE))
kommun <- left_join(kommun_peryear, kommun_tot, by="Kommun")
```

```r
head(kommun)
```

**Vilken kommun har fått mest i LONA-bidrag?**

**Summera per kommun och år.**

```r
topplista <- arrange(kommun, desc(`Totalt beviljat (kr)`))
```

```r
head(topplista)
```
